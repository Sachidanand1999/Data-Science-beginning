# Creating a Boxplot

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
cars = pd.read_csv(r"E:\Python Task\Cars.csv")
```

```python
cars
```

```python
cars.plot(kind = 'box', subplots = True, sharey = False, figsize = (10,8))
plt.subplots_adjust(wspace = 0.75)
plt.tight_layout()
```
