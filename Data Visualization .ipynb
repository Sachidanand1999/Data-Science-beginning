### Heatmap

```python
corr_mat = cars.corr( method = 'pearson' )
```

```python
print(corr_mat)
```

```python
sns.heatmap( corr_mat, xticklabels = corr_mat.columns, yticklabels = corr_mat.columns, cmap = 'coolwarm' )
plt.tight_layout()
plt.title( 'Heatmap of cars dataset' )
plt.show()
```
